La presente libreta de Jupyter tiene como propósito el analisis del registro de género por escuela dentro de la entidad federativa de Jalisco en México, dicho analisis se basa en datos recopilados por la Secretaria de Educación en conjunto con la Inegi.

Link de descarga del archivo .csv: https://datos.jalisco.gob.mx/dataset/cct-estadistica 

```python
#Se importan las siguientes librerias con el fin de analizar el archivo .csv y poder graficar sus resultados
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#kLectura del csv mediante la creaci+on de un pandas dataframe
df = pd.read_csv("alumnos.csv", encoding = 'latin-1')
```

```python
#Verificación de la cantidad de datos que tenemos, así como su tipo de variable
df.info()
print("Numero de datos: " , len(df))
```

## Resumen Estadístico

```python
#Informe sobre los principales datos estadísticos de cada una de las columnas del dataframe
df.describe()
```

## Rangos

```python
#Impresión del rango de cada una de las columnas del DataFrame
for col in df.columns:
    try:
        print(col, ':',df[col].min(),  "-", df[col].max())
    except:
        print("", end = "")  
```

## Media

```python
#Impresión de la media de cada una de las columnas del DataFrame
for col in df.columns:
    try:
        print('Media', col, ':' ,df[col].mean())
    except:
        print("", end = "")  
```

## Mediana

```python
#Impresión de la mediana de cada una de las columnas del DataFrame
for col in df.columns:
    try:
        print('Mediana', col, ':' ,df[col].median())
    except:
        print("", end = "")  
```

## Desviación estándar

```python
#Impresión de la desviación estándar de cada una de las columnas del DataFrame
for col in df.columns:
    try:
        print('std', col, ':' ,df[col].std())
    except:
        print("", end = "")  
```


### A continuación se presentan unas gráficas que tienen como fin ilustrar los datos presentados anteriormente

```python
fig = plt.figure(figsize=(6, 4))

sns.countplot(data=df, y = 'NOM TURNO')

plt.title('Turnos de los Alumnos')
plt.xlabel('Cantidad de alumnos')
plt.ylabel('Tipo')
```

```python
fig = plt.figure(figsize=(6, 4))

sns.countplot(data=df, y = 'NIVEL')

plt.title('Nivel de Educación')
plt.xlabel('Cantidad de alumnos')
plt.ylabel('Tipo')
```

## Conclusiones de los datos

En promedio el número de hombres y mujeres por grado es casi igual, tiene pequeñas variaciones de hasta 1 alumno por grado.

La mediana es el punto medio del conjunto de datos. El valor de este punto medio es el punto en el cual la mitad de las observaciones esta por encima del valor y la otra mitad esta debajo de este. Con base en lo anterior podemos destacar, por ejemplo, que existen casos en de salones en donde existe menos de 4 alumnos independienetemente de su género, como es el caso de los alumnos de primer grado.

Una desviación estandar baja indica que la mayor parte de los datos de una muestra tiende a estar agrupados cerca del valor de la media, mientras que una desviación estandar alta indica que los valores se extienden sobre un rango de valores mas amplio, en este caso podemos observar que la desviacion estandar llega a tener un valor de hasta 52 alumnos en el total de alumnos de 1°, por lo que consideramos eso como una desviación estandar, esto quiere decir que el numero de alumnos de cada salon llega a variar mucho entre cada escuela.
