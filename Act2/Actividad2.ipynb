```python
#Se importan las siguientes librerias con el fin de analizar el archivo .csv y poder graficar sus resultados
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Lectura del csv mediante la creaci+on de un pandas dataframe
df = pd.read_csv("alumnos.csv", encoding = 'latin-1')
```

```python
fig = plt.figure(figsize=(6, 4))

sns.countplot(data=df, y = 'NOM TURNO')

plt.title('Turnos de los Alumnos')
plt.xlabel('Cantidad de alumnos')
plt.ylabel('Tipo')
```

```python
fig = plt.figure(figsize=(6, 4))

sns.countplot(data=df, y = 'NIVEL')

plt.title('Nivel de Educación')
plt.xlabel('Cantidad de alumnos')
plt.ylabel('Tipo')
```

```python
# Tamaño de la imagen
fig = plt.figure(figsize=(10, 5))

# Gráfico boxplot
sns.boxplot(data=df, x='HOMBRES 1¡')

# Ejes y título
plt.title('Mapa de cajas')
```

```python
# Tamaño de la imagen
fig = plt.figure(figsize=(25, 25))

# Gráfico boxplot
sns.boxplot(data=df, x='DOCENTES')

# Ejes y título
plt.title('DOCENTES DE IDIOMAS')
```
